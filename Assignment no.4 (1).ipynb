```python
#Quiz application
questions = [
    {
        "s no." : 1,
        "question" : "what is your country name",
        "options" : ["pakistan","usa","uae","africa"],
        "answer" : "pakistan"
    },
    {
        "s no." : 2,
        "question" : "what is the captial of pakistan",
        "options" : ["islamabad","karachi","murree","lahore"],
        "answer" : "islamabad"
    },
    {
        "s no." : 3,
        "question" : "what is your city name",
        "options" : ["karachi","lahore","islamabad","peshawar"],
        "answer" : "karachi"
    },
    {
        "s no." : 4,
        "question" : "what is your name",
        "options" : ["sajjad","Zubair","huzaifa","taimoor"],
        "answer" : "Sajjad"
    },
    {
        "s no." : 5,
        "question" : "what is your father name",
        "options" : ["rafat","sultan","aman ullah","yousuf"],
        "answer" : "Rafat"
    },
]


score = 0

for question in questions:
    data = str(question["s no."]) + ": " + question["question"] + "\n please write correct answer in input field \n" + question["options"][0] + "\n" + question["options"][1] + "\n" + question["options"][2] + "\n" + question["options"][3] + "\n"
    
    answer = input(data)
    if answer.lower() == question["answer"]:
        score += 1

percentage = ((score/5)*100);
totalQuestion = len(questions)

if(percentage > 50 and percentage< 80):
    print("\n\nCongratulations: you pass with average percentage of " + str((score/5)*100)+"%")    
elif(percentage >= 80 and percentage<= 100):
    print("\n\nCongratulations: you pass with excelent percentage of " + str((score/5)*100)+"%")    
elif(percentage < 50):
    print("\n\nSo sorry, you're fail with percentage of "+ str((score/5)*100)+"%")


# print(percentage)
print("\nyour correct answer is :" + str(score)+" out of "+str(totalQuestion))
```
